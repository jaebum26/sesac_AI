## [1] 파이썬 모듈 이해하기
- 모듈이란 함수나 변수 또는 클래스를 모아 놓은 파이썬 파일이다
- 모듈은 다른 파이썬 프로그램에서 불러와 사용할 수 있도록 만든 파이썬 파일이라고도 할 수 있다


```python
####################################################
# guguDan.py 내용임
# 함수를 미리 실행하여 봅니다.  def 함수명(): 파일
####################################################

def gugu():
    x=input('출력할 구구단의 단을 선택해주세요. 1~9')
    x=int(x)
    for num in range(9):
        print('%d*%d=%d'%(x,num,x*num))

gugu() # 실행한후 숫자를 입력합니다. 
```

```python
class prn():
    def line(x='-',y=50):
        print(x*y)

    def ti():
        print('''
      이 프로그램은
      파이썬에서 작성
      
      ''')

    def gu():
        print(1*1,end=(", "))
        print(1*2,end=(", "))
        print(1*3)

prn.line()
prn.ti()
prn.line()
prn.gu()
prn.line()
```

```python
####################################################
# __init__.py 파일 내용임
####################################################
# import guguDan  # 이 작업은 __init__.py 에서만 실행가능함으로 주석처리함

def title():
    print(''' ver: python 3.8 
              year: 2023
        ''')

def line():
    print('-'*20)
    
def bu():
    line()
    title()
    line()
    print('이번과정은 파이썬의 모듈 연습입니다.')
    line()
	
bu()
```

```python
####################################################
# test 폴더를 import 하면 __init__.py 를 자동으로 실행함
####################################################

import test                 # __init__.py 의 함수 힐행
import test.guguDan as b    # __init__.py 에서 import guguDan 된 모듈을 b 로 별명
test.line()
test.title()

test.line()
b.gugu()    # 입력할 단숫자를 넣음.
```

## [3] 파이썬 메모리 제거
- del 변수  로 일부 변수만 제거
- 아래의 명령으로 메모리 모든 내용 제거


```python
for name in list(globals().keys()):
    if not name.startswith('_'):
        del globals()[name]
```
