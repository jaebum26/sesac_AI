```python
####################################################
# guguDan.py 
#    .  def (): 
####################################################

def gugu():
    x=input('   . 1~9')
    x=int(x)
    for num in range(9):
        print('%d*%d=%d'%(x,num,x*num))

gugu() #   . 
```

```python
class prn():
    def line(x='-',y=50):
        print(x*y)

    def ti():
        print('''
       
       
      
      ''')

    def gu():
        print(1*1,end=(", "))
        print(1*2,end=(", "))
        print(1*3)

prn.line()
prn.ti()
prn.line()
prn.gu()
prn.line()
```

```python
####################################################
# __init__.py  
####################################################
# import guguDan  #   __init__.py   

def title():
    print(''' ver: python 3.8 
              year: 2023
        ''')

def line():
    print('-'*20)
    
def bu():
    line()
    title()
    line()
    print('   .')
    line()
	
bu()
```
