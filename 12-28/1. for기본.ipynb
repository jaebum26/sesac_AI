```python
file=r"C:\Users\blucom005\Downloads\12-27\12-27\img\flower03.jpg"

from PIL import Image
img = Image.open(file)
img.size #너비 416, 높이 310    이 자료는 행이 310개, 열이 416개
print(type(img))
```

```python
###################################
### 이미지를 벡터화 함
### 인공지능은 벡터자료만 취급
### PIL에서 제공되지 않은 작업 중 내 맘대로 이미지 변형하고자 할 때
###################################


import numpy as np
imgArr=np.array(img)
print(type(imgArr))
imgArr.shape

tmp = imgArr[:100,:,:] #0번행에서 99번행까지 100개 행의 자료 중, 모든 열 모든 RGB를 다 보여줘
tmp = imgArr[:,30:50,:] #모든 행과 30:49번까지의 열, 모든 RGB를 보여줘
imgArr[100:,:,:]=0
```

```python
import matplotlib.pyplot as plt
plt.imshow(imgArr)
```
