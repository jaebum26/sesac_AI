```python
import tkinter as tk
from tkinter import messagebox
import sqlite3
from datetime import datetime

def login():
    username = entry_username.get()
    password = entry_password.get()

    # SQLite 데이터베이스 연결
    conn = sqlite3.connect('database.db')
    cursor = conn.cursor()

    # 사용자 정보 조회
    query = "SELECT * FROM customers WHERE customerName=? AND customerPass=?"
    cursor.execute(query, (username, password))
    user = cursor.fetchone()

    # 로그 기록
    log_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    if user:
        log_query = "INSERT INTO logs (username, login_time, success) VALUES (?, ?, ?)"
        cursor.execute(log_query, (username, log_time, 1))
        conn.commit()
        messagebox.showinfo('로그인 성공', '환영합니다, {}'.format(username))
    else:
        log_query = "INSERT INTO logs (username, login_time, success) VALUES (?, ?, ?)"
        cursor.execute(log_query, (username, log_time, 0))
        conn.commit()
        messagebox.showerror('로그인 실패', '잘못된 사용자 이름 또는 비밀번호입니다.')

    conn.close()

# SQLite 데이터베이스 초기화
conn = sqlite3.connect('database.db')
cursor = conn.cursor()

# logs 테이블 생성
cursor.execute('''
    CREATE TABLE IF NOT EXISTS logs (
        logID INTEGER PRIMARY KEY,
        username TEXT,
        login_time TEXT,
        success INTEGER
    )
''')

conn.commit()
conn.close()

# GUI 생성
root = tk.Tk()
root.title('로그인 화면')

# 레이블 및 입력 필드 추가
label_username = tk.Label(root, text='사용자 이름:')
label_username.pack()

entry_username = tk.Entry(root)
entry_username.pack()

label_password = tk.Label(root, text='비밀번호:')
label_password.pack()

entry_password = tk.Entry(root, show='*')  # 비밀번호를 '*'로 표시
entry_password.pack()

# 로그인 버튼 추가
btn_login = tk.Button(root, text='로그인', command=login)
btn_login.pack()

# GUI 실행
root.mainloop()
```
