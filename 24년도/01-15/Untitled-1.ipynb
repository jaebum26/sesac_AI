```python
import pandas as pd
import folium

# 엑셀 파일에서 데이터 읽기
happy_data = pd.read_excel('happy.xlsx')

# Folium 지도 생성
map_happy = folium.Map(location=[36.5, 127.5], zoom_start=7)

# 데이터를 기반으로 원을 추가하여 지도에 표시
for index, row in happy_data.iterrows():
    location = (row['longitude'], row['latitude'])
    happiness_score = row['건강수치']

    # 원의 크기를 건강수치에 따라 조절
    radius = happiness_score / 10  # 예시 값, 조절 가능

    folium.CircleMarker(
        location=location,
        radius=radius,
        popup=f'Happiness Score: {happiness_score}',
        color='green',
        fill=True,
        fill_color='green'
    ).add_to(map_happy)

# 생성된 지도를 저장하거나 표시
map_happy.save('happy_map.html')
map_happy
```
