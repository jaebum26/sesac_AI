```python
## 엑셀자료 복사붙임
## 파이썬에서 두줄이상의 문자열을 처리할때 '''     '''
data='''22.2	14.5	11.6	14.7	16.2	17.5	15.3	18.3	8.4	14.5
7.4	13.7	11.7	18.7	14.1	12.5	13.1	7.7	7.5	17.2
20.5	9.4	15.5	11.5	10.6	19.4	5.5	10.7	20	20.4'''
data
```

- 탭키와 줄바꿈(\n)이 섞여 있어서 한개의 키로 통일하고자함.


```python
# gpt --> 파이썬 문자열 바꾸기

data = data.replace("\n", "\t")
data
```

```python
# gpt--> 파이썬 문자열 나누기
data = data.split('\t')
data
```

#### 컴프리헨션
- 컴프리헨션은 for를 3줄 작업을 한줄에 압축한 기술임
- num_list=[]
- for x in data:
    - num_list=float(x)

```python
## gpt --> 문자열 리스트 숫자화
num_list = [float(x) for x in data]   
num_list
```

```python
import seaborn as sns
sns.set_theme(style='whitegrid')
ax = sns.histplot(x=num_list,
                  color='r', ## 색상
                  bins=10, ## bin 개수
                  kde=True, ## 밀도 함수 곡선 추가
                  element='step', ## 히스토그램 표시형식,
                  stat='percent', ## y값 계산 방식 count
                  cumulative=False, ## True인 경우 누적 분포 형태로 그림
                 )
```

```python
import matplotlib.pyplot as plt
plt.hist(num_list,bins=10)
```

```python
#gpt -->  문자열리스트를 시간으로

time_strings=new_word_list
from datetime import datetime

# 문자열을 시간으로 변환하는 함수
def convert_to_datetime(time_str):
    return datetime.strptime(time_str, "%Y-%m-%d %H:%M")

# 리스트 내의 문자열들을 시간으로 변환
time_list = [convert_to_datetime(time_str) for time_str in time_strings]
time_list[:5]
```
