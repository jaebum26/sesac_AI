```python
## 작업3

import csv

data = []
f = open('울산광역시 남구_코로나19_확진자_및_사망자_현황_20220126.csv','r',encoding='euc-kr')
reader = csv.reader(f)

for row in reader:
    data.append(row)
f.close()
```

```python
## 작업4

header=data[0] ; local=data[1:]
label=[]

for x in local:
    label.append(x[2]+'-'+x[3])

num=[]

for x in local:
    num.append(x[4:6])
```

```python
## 작업5

#### num에 있는 자료가 문자로 들어가 있음.
new_num=[]

for x in num:
    line=[]
    for y in x:
        line.append(int(y))
    new_num.append(line)
new_num
```

```python
## 작업6

#############################################
############# numpy로 합
#############################################
import numpy as np

세로합=np.sum(new_num, axis=0)
세로합

import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic') #NanumBarunGothic 나눔바른고딕 글꼴이 없다고 해서 맑은 고딕으로 썼습니다.
plt.bar(header[4:6],세로합)
plt.show()
```

```python
## 작업7

세로합_2020년도=np.sum(new_num[:9],axis=0)
세로합_2021년도=np.sum(new_num[9:21],axis=0)

print(세로합_2020년도)
print(세로합_2021년도)
```

```python
## 작업8

plt.figure(figsize=(20,5))
plt.subplot(121)
plt.title('2020년도 확진자, 사망자 수')
plt.bar(header[4:6],세로합_2020년도)
#plt.show()

plt.subplot(122)
plt.title('2021년도 확진자, 사망자 수')
plt.bar(header[4:6],세로합_2021년도)
plt.show()
```
