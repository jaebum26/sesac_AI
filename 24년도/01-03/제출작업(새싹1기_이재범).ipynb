```python
## 3

import csv

data = []
f = open(' _19_____20220126.csv','r',encoding='euc-kr')
reader = csv.reader(f)

for row in reader:
    data.append(row)
f.close()
```

```python
## 4

header=data[0] ; local=data[1:]
label=[]

for x in local:
    label.append(x[2]+'-'+x[3])

num=[]

for x in local:
    num.append(x[4:6])
```

```python
## 5

#### num     .
new_num=[]

for x in num:
    line=[]
    for y in x:
        line.append(int(y))
    new_num.append(line)
new_num
```

```python
## 6

#############################################
############# numpy 
#############################################
import numpy as np

=np.sum(new_num, axis=0)


import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic') #NanumBarunGothic       .
plt.bar(header[4:6],)
plt.show()
```
