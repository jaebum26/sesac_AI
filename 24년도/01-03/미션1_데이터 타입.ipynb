### [미션2] 데이터 원소갯수 구하기 (len함수)

```python
len(data)
```

```python
# 미션2-1: 여행지, 예상경비1일, 예상일정의 항목갯수(원소갯수)를 print문으로 출력합니다.

print(len(data['여행지']),len(data['예상경비1일']),len(data['예상일정']))
```

```python
# 미션2-2 : for를 이용하여서 미션 2-1 구문을 변경합니다. 
itemList=['여행지','예상경비1일','예상일정']

for a in itemList:
    print(len(data[a]))
```

```python
# 미션2-3 : 미션 2-1의 for구문에 if문을 추가합니다.
# 항목갯수가  3미만(<3)일때는 print문으로 '항목갯수가 작음. 추가바람'
itemList=['여행지','예상경비1일','예상일정']

for a in itemList:
    if len(data[a]) >= 3:
        print(len(data[a]))
    else:
        print('항목개수가 작음.추가바람')
```

### [미션3] 랜덤값으로 여행일정 짜기

```python
import random
randomNum=random.randrange(0,4)   # 0,1,2,3의 random값 나옴 ctrl+enter 여러번 수행하면서 확인
randomNum
```

```python
## 미션 3-1: 여행지의 갯수가 0,1,2,3,4번 5개임, 0~5개의 숫자가 나올수 있도록 작성합니다.
##  여행지에서 random하게 지역을 출력합니다.

print(randomNum)

selectedTrip = random.choice(ftrip)
print(selectedTrip)
```

```python
## 미션 3-2: random으로 나올수 있는 마지막 값을 여행지의 갯수 len함수로 지정함.

sTrip = len(ftrip)
sDate = len(fdate)
sMoney = len(fmon)

selectedTrip = random.randrange(sTrip)
print(ftrip[selectedTrip-1])
```

```python
## 미션 3-3: random으로 여행지, 예상경비1일,예상일정을 출력하세요

selectedTrip = random.randrange(sTrip)
print(ftrip[selectedTrip-1])
selectedMoney = random.randrange(sMoney)
print(fmon[selectedMoney-1])
selectedDate = random.randrange(sDate)
print(fdate[selectedDate-1])
```

```python
## 미션 3-4: 미션3-3을 for 로 작성하지 않았다면 for로 변경
def randTrip(code,key):
    selectedTrip = random.randrange(code)
    print(key[selectedTrip-1])

for key, value in data.items():
    randTrip(len(value), value)
```

```python
## 미션 3-5: for로 변경된 구문울 이용하여서  랜덤여행일정 리스트를 제작
# 출력예 ==> 랜덤여행일정=['경기','50000','3일']

randomTripList = []

for key, value in data.items():
    selected_item = random.choice(value)
    randomTripList.append(selected_item)

print('랜덤여행일정 =',randomTripList)
```

```python
# 미션3-6: 랜덤여행일정의 자료의 1번 인덱스의 '50000'을 숫자형으로 변경
#          랜덤여행일정 자료의 2번 인덱스의 '3일'의 '일'을 제거하고 숫자형으로 변경 (구글검색  '파이썬 replace')
#          최종결과는
# 랜덤여행일지=['경기',50000,3,150000]    #마지막은 1일경비*예상일정

```
