#### ascii코드 참고(이코드 가장 하단에 코드있음)


```python
##  좌표이동 https://diytube.tistory.com/21

import cv2
img_color=cv2.imread('sample.jpg', cv2.IMREAD_COLOR)  

x, y = 200, 200
title = 'KeyBoard Control'
img=img_color.copy()

while True:
    cv2.imshow(title, img)
    cv2.moveWindow(title, x, y)
    key = cv2.waitKey(0) & 0xFF # 키보드 입력 대기, 8비트 마스크 처리
    print(key, chr(key))        # 키보드 입력 값, 문자 값 출력
    if key == ord('a'):
        x -= 10                 # a가 입력되면 왼쪽으로 이동
    
    elif key == ord('s'):
        y += 10                 # s가 입력되면 아래로 이동
    
    elif key == ord('w'):
        y -= 10                 # d가 입력되면 위로 이동
        
    elif key == ord('d'):
        x += 10                 # w가 입력되면 오른쪽으로 이동
    
    elif key == ord('q') or key == 27:
                 
        cv2.destroyWindow(title) # q이거나 esc이면 종료.
        break   
    cv2.moveWindow(title, x, y) # 새로운 좌표로 창 이동
```

### <div class="alert alert-block alert-info"> [Part3] openCV에서 이미지 cv2.COLOR_BGR2RGB
- 오픈CV는 이미지를 읽을때 rgb를 bgr로 배열함
- 오픈cv의 cv2.imshow 로 출력할때는 openCV에서 자체적으로 rgb로 다시 변환해주시만 plt.imshow로 볼때는 rgb가 아닌 bgr로 출력됨으로
  사용자가 rgb로 변환하여야 함.

```python
###############################
### 
##############################
import cv2
from PIL import Image
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'  #한글깨짐해결
plt.rcParams['axes.unicode_minus'] =False  # 마이너스 기호 깨짐 방지

plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
img=Image.open('sample.jpg')
plt.imshow(img)
plt.title('rgb이미지, 정상이미지') 

plt.subplot(1,3,2)
img=cv2.imread('sample.jpg')
plt.imshow(img)
plt.title('bgr이미지')


plt.subplot(1,3,3)
img=cv2.imread('sample.jpg')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('openCVrgb이미지, 정상이미지')
```

```python
###########################################
### (참고) 사용자가 직접 rgb의 위치값을 변경할수도 있음. 
### 이러한 작업을 swap라함 (코딩테스트용)
### swap를 이용한 bgr를 rgb로 변경
#############################################
img=cv2.imread('sample.jpg')   # BGR로 읽힘

class my():
    def BgrRGB(img):
        tmp=img.copy()      #파이썬에선 .copy를 사용하여 swap을 해줌
        img[:,:,0]=tmp[:,:,2]
        img[:,:,2]=tmp[:,:,0]
        
        return img

img=my.BgrRGB(img)

plt.imshow(img)
```

```python
########################################
### 함수제작
#########################################

import cv2
import matplotlib.pyplot as plt
import numpy as np

def imgRead(file,option=cv2.IMREAD_COLOR): #option=1
    img=cv2.imread(file,option)
    return img

def chart(imgList,option):
    plt.figure(figsize=(10,3))
    for cnt,file in enumerate(imgList):
        img=imgRead(file,option)
        plt.subplot(1,len(imgList),cnt+1)
        plt.imshow(img)
        plt.title('fileName: {0}, \n shape: {1})'.format(file,np.shape(img)))
        
imgList=['sample.jpg','pen.png','navi.png']       
chart(imgList,-1)
```

### <div class="alert alert-block alert-info"> [Part5] 이미지의 관심영역(ROI)에 대한 처리
-  ROI: (Region of Interest, ROI: 이미지 내에서 관심 영역 )



```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('dog.jpg',cv2.IMREAD_COLOR)   # rgb 3채널로 읽음
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)    # plt.imshow로 보기 위하여서 bgr를 rgb로 변환

plt.rcParams['font.family'] ='Malgun Gothic'  #한글깨짐해결
plt.rcParams['axes.unicode_minus'] =False  # 마이너스 기호 깨짐 방지

plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
plt.imshow(img)
plt.title('원본이미지')

plt.subplot(1,3,2)
roi=img[400:600, 250:500]
plt.imshow(roi)
plt.title('행(400:600), 열(250:500)')

plt.subplot(1,3,3)
roi=img[600:, 500:]
plt.imshow(roi)
plt.title('행(600이후), 열(500이후)')
```

###  5-1 이미지의 일정 영역 검정색으로

```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('dog.jpg',cv2.IMREAD_COLOR)   # rgb 3채널로 읽음
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)    # plt.imshow로 보기 위하여서 bgr를 rgb로 변환


plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
plt.imshow(img)
plt.title('원본이미지')

plt.subplot(1,3,2)
img[400:600, 250:500]=0   # 이 영역에 흰색(0,0,0)으로 변경
plt.imshow(img)
plt.title('행(400:600), 열(250:500)')

plt.subplot(1,3,3)
img[600:, 500:]=255      # 위에 일정영역은 흰색으로 변경되었고, 또 600:,500: 이후는 검정색 255로 변경됨
plt.imshow(img)
plt.title('행(600이후), 열(500이후)')
```

#### [5-4]이미지를 네모 영역으로 그리기

```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('dog.jpg',cv2.IMREAD_COLOR)   # rgb 3채널로 읽음
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   


### 0,200 은  시작하는 왼쪽 위치가 0, 시작하는 위쪽 위치가 200 (즉 시작하는 x,y 가 0,200)
### 700,500은 끝나는 오른쪽 위치가 700, 끝나는 아래쪽 위치가500 (즉 끝나는 x,y 가  700,500)
tmp=cv2.rectangle(img,(0,200),(700,500),(255,0,0),3)
plt.imshow(tmp)
```

#### [5-5]이미지를 다각형 영역으로 작업하기
- 다각형은  각점(행,열 위치), 각점....을 연결하는 도형을 의미함.
    - https://csshark.tistory.com/191

```python
import numpy as np
img=np.zeros((20,20,3),np.uint8)

plt.figure(figsize=(10,3))
point=np.array([[3,1],[12,5],[15,17]], np.int32)  # 3열1행(왼쪽에서3번째,위에서1), 12열5행, 15열17행

plt.subplot(1,2,1)
canvas=cv2.polylines(img,[point],False,(255,0,0))
plt.imshow(canvas)
plt.title('열린 폴리선')


plt.subplot(1,2,2)
canvas=cv2.polylines(img,[point],True,(255,0,0))
plt.imshow(canvas)
plt.title('닫힌 폴리선')

```

```python
##################################
### 아래 이미지에서 빨간색 선 도형
####################################
import numpy as np
img=cv2.imread('sample.jpg',cv2.IMREAD_COLOR)   # rgb 3채널로 읽음
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)    # plt.
point=np.array([[50,100],[40,110],[30,130],[0,250],[200,250],[180,200],[50,100]], np.int32)

plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
aa=cv2.polylines(img,[point],True, (255,0,0))
plt.imshow(aa)

plt.subplot(1,2,2)
aa=cv2.fillConvexPoly(img,point,(255,0,0))
plt.imshow(aa)


```
