```python
from collections import Counter

network_list = []
for networks in drama['Network'].to_list():
    networks = networks.strip().split(", ")
    for network in networks:
        network_list.append(network)
        
network_df = pd.DataFrame.from_dict(Counter(network_list),orient='index').rename(columns={0:'Count'})
network_df.sort_values(by='Count',ascending = False,inplace = True)
network_df
```

# 데이터 가공 및 그룹핑

```python
# 방영일(Aired Date), 에피소드 길이(Duration),줄거리(Synopsis),출연자(Cast),태그(Tags)
col=['Aired Date','Duration','Synopsis','Cast','Tags']
newdrama=drama.drop(col,axis=1)
newdrama.columns
```

```python
# 영문컬럼명을 아래와 같이 한글컬럼명으로 변경한다.
newdrama.columns=['이름','개봉년도','요일','에피소드수','네트워크','등급','장르','순위','평점']
newdrama.columns
```

```python
newdrama['장르개수']=newdrama['장르'].str.count(',')+1
newdrama.head()
```

```python
newdrama['네트워크개수']=newdrama['네트워크'].str.count(',')+1
newdrama.head()
```

## 데이터 그룹핑

```python
# 개봉년도별 드라마 개수
d=newdrama.groupby('개봉년도')['이름'].count()
pd.DataFrame(d)
```

```python
# (개봉년도, 평점) 별 드라마 개수
d=newdrama.groupby(['개봉년도','평점'])['이름'].count()
pd.DataFrame(d).head(10)
```

```python
d=newdrama.groupby('개봉년도')['평점'].mean()
pd.DataFrame(d)
```

```python
from collections import Counter
newdrama['장르'] = newdrama['장르'].str.strip()
genre_list = list()
for genres in newdrama['장르'].to_list():
    genres = genres.split(",  ")
    for gen in genres:
        genre_list.append(gen)       
genre_df = pd.DataFrame.from_dict(Counter(genre_list), orient = 'index').rename(columns = {0:'Count'})
genre_df.sort_values('Count',ascending = False, inplace = True)
genre_df.head()
```

```python
plt.figure(figsize=(25,10))
plt.plot(genre_df,color='r',marker='o')
plt.bar(genre_df.index,genre_df.Count)
plt.title('장르별 드라마 개수')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```
