###문제9

```python
month = int(input('월 입력: '))
if month < 1 or month > 12 :
    print('존재하지 않는 월!!')
elif 3 <= month <= 5 :
    print(f'{month}월은 봄')
elif 6 <= month <= 8 :
    print(f'{month}월은 여름')
elif 9 <= month <= 11 :
    print(f'{month}월은 가을')
else :
    print(f'{month}월은 겨울')
```

###문제11

```python
import random
com = random.randint(1,20)
print('<< 컴퓨터가 생각한 1~20 숫자 맞히기 >>')
while True :
    player = int(input('숫자 입력(종료 0): '))
    if player == 0 :
        break
    elif player == com :
        print('정답!!')
        break
    elif player > com :
        print('더 작은 숫자 입력! ')
    elif player < com :
        print('더 큰 숫자 입력! ')
```

###문제12

```python
import random
word = ['잘 지냈지?','조금은','어색해','그리운','목소리','기다림의','끝은','기적이','되고','소중해']
input('타자게임 시작 (엔터 입력) ')
w = random.choice(word)
n = 1        # 문제번호
while True :
    print('문제',n,'(종료 0):',w)
    my = input()
    if my == '0' :
        break
    elif my == w :
        print('맞음!!\n')
        w = random.choice(word)
    else :
        print('틀림! 다시!\n')
    n += 1    # 문제번호 증가
```

###문제14

```python
def price(menue):
    if menue==1:
       m="아메리카노"
       p="2500원"
    elif menue==2:
        m="카페라떼"
        p="3000원"
    elif menue==3:
        m="카푸치노"
        p="3000원"
    print(m,p)


menue=int(input('메뉴선택(1:아메리카노/2:카페라떼/3:카푸치노) '))
price(menue)
```
