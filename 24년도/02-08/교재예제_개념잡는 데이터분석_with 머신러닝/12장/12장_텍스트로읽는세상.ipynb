```python
import collections
text='이승윤 숨고 싶을 땐 다락이 되어 줄거야 죽고 싶을 땐 나락이 되어 줄거야 울고 싶은 만큼 허송 세월 해 줄거야 진심이 버거울 땐 우리 가면 무도회를 열자'
sentece_tag=okt.pos(text)
sentece_tag
adj_list=[]
for word,tag in sentece_tag:
  if tag in ['Noun','Adjective']:
    adj_list.append(word)
counts=collections.Counter(adj_list)
tag=counts.most_common(2)
tag
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import collections
from wordcloud import WordCloud
```

```python
df=pd.read_excel('/content/12장_실습2.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['정규화본문'] = df['본문'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 0-9]","")
df.head()
```

```python
temp_list=[]
for sentence in df['정규화본문']:
 s_list=okt.pos(sentence)
 for word,tag in s_list:
   if tag in ['Noun','Adjective']:
     temp_list.append(word)
counts=collections.Counter(temp_list)
tag=counts.most_common(50)
tag
```

```python
from wordcloud import WordCloud
wc=WordCloud(font_path=font_path,background_color='white', max_font_size=60, scale=2.0)
cloud=wc.generate_from_frequencies(dict(tag))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.imshow(cloud)
```

```python
import numpy as np
from PIL import Image
img = Image.open('/content/검은배경.jpg')
img_array = np.array(img)

wc = WordCloud(font_path=font_path, width=400, height=400, scale=2.0, max_font_size=250, mask=img_array)
gen = wc.generate_from_frequencies(dict(tag))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.imshow(gen)
```

연습문제5

```python
temp_list=[]
stopword=['중','기자','이','수','위','등','것','있다','날','씨','위해','원','개','입니다','를','제','의','고','명','있는','전','두','관','윤','파','시']
for sentence in news_df['정규화본문']:
 s_list=okt.pos(sentence)
 for word,tag in s_list:
   if word not in stopword:
    if tag in ['Noun','Adjective']:
      temp_list.append(word)
counts=collections.Counter(temp_list)
tag=counts.most_common(5)
tag
```

연습문제6

```python
tag_200=counts.most_common(200)
wc=WordCloud(font_path=font_path,background_color='white', max_font_size=60, scale=2.0)
cloud=wc.generate_from_frequencies(dict(tag_200))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.imshow(cloud)
```
