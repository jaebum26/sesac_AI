# 데이터 가공하기

```python
# 데이터불러오기
import pandas as pd
filename='/content/drive/MyDrive/music.xlsx'
df=pd.read_excel(filename)
df.head()
```

## 컬럼(변수) 삭제하기

```python
# 데이터 분석에 필요없는 publishing_company(발매사), lyricist(작사가), composer(작곡가)를 삭제
df.drop(['publishing_company','lyricist','composer'], axis=1, inplace=True)
```

```python
df.columns
```

## 컬럼 이름 변경하기

```python
# 모든 컬럼 이름 변경하기
df.columns=['제목','이름','성별','인원수','구분','발매일','기획사','장르','유형']
df.head()
```

## 컬럼 생성하기

```python
df.dtypes
```

```python
# 새로운 컬럼 생성하기 : 년도,월,일
df['년도']=df['발매일'].dt.year
df['월']=df['발매일'].dt.month
df['일']=df['발매일'].dt.day
df.head()
```

```python
df
```

## 데이터 그룹핑

```python
# 장르별 가수인원수 출력하기
newdata=df.groupby('장르').count()
pd.DataFrame(newdata['인원수'])
```

```python
# 구분,장르별  가수 인원 수 합계
newdata=df.groupby(['구분','장르']).sum()
pd.DataFrame(newdata['인원수'])
```

```python
# 다음 tips 데이터는 미국 레스토랑의 매출과 웨이터에게 지불한 팁을 나타내는 데이터이다. 
# 아래의 코드와 실행결과를 참고하여 다음 문제를 차례대로 해결해보자.
import pandas as pd
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
# 2-1) 다음은 전체 데이터 중 “팁이 7달러 이상인” 데이터를 추출하는 코드이다. 빈 칸에 가장 알맞은 코드를 완성하시오. 
tips[tips['tip']>=7]
```

```python
tips[tips['sex']=='Female']
tips[tips['smoker']=='Yes']
```

```python
# 2-2) 다음은 “여자손님이면서 비흡연자인” 데이터를 추출하는 코드이다. 빈 칸에 가장 알맞은 코드를 완성하시오.
conditions=(tips['sex']=='Female') & (tips['smoker']=='Yes')
tips[conditions].head()
```

```python
# 2-3) 다음은 “팁을 가장 많이 지불한 고객”에 대한 데이터 추출하는 코드이다. 빈 칸에 가장 알맞은 코드를 완성하시오. 
tips[tips['tip'].max()==tips['tip']]
```

```python
#2-4) 다음은 total_bill과 tip에 대한 내림차순 정렬을 한 결과 상위 10개만 출력하는 코드이다. 빈 칸에 가장 알맞은 코드를 완성하시오. 
tips.sort_values(['total_bill','tip'],ascending=False).head(10)
```

```python
# 2-5) 다음은 size를 기준으로 내림차순, tip을 기준으로는 오름차순 정렬한 결과 상위 10개의 데이터만 출력하는 코드이다. 빈 칸에 가장 알맞은 코드를 완성하시오. 

tips.sort_values(['size','tip'],ascending=[False,True]).head(10)
```

```python
# 2-6) 다음은 total_bill과 tip의 합계인 total 컬럼을 생성하는 코드이다. 빈 칸에 가장 알맞은 코드를 완성하시오. 
tips['total']=tips['total_bill']+tips['tip']
tips
```

```python
# 2-7) 다음은 day별 tip의 평균을 출력하는 코드이다. 빈 칸에 가장 알맞은 코드를 완성하시오. 
pd.DataFrame(tips.groupby('day').mean()['tip'])
# 일요일에 팁 평균이 가장 높았음을 알 수 있다.
```

```python
# 문제 3. 문제2번의 tip데이터셋을 이용하여 아래의 문제를 차례대로 해결해보시오.
# 1) tips 데이터셋 중 day가 금요일(Fri), 토요일(Sat)만 필터링한다.
# 2) tip이 평균보다 많게 낸 데이터만 필터링한다.
# 3) 컬럼은 total_bill, tip, smoker, time만 출력한다.
# 4) total이 높은 상위 10개 행만 출력한다.
```

```python
# 1) tips 데이터셋 중 day가 금요일(Fri), 토요일(Sat)만 필터링한다.
conditions=(tips['day']=='Fri') | (tips['day']=='Sat')
#conditions=tips['day'].isin(['Fri','Sat']) # 특정 조건 선택 시 isin함수를 많이 사용한다.
tips[conditions]
```

```python
#2) tip이 평균보다 많게 낸 데이터만 필터링한다.
tips[tips['tip']>=tips['tip'].mean()]
```

```python
# 3) 컬럼은 total_bill, tip, time, total만 출력한다.
tips[['total_bill','tip','time','total']]
```

```python
# 4) total이 높은 상위 10개 행만 출력한다.
tips.sort_values('total',ascending=False).head(10)
```
