## 4.2 맷플롯립(Matplotlib) 라이브러리

```python
import matplotlib.pyplot as plt
plt.title('Line Graph')
xdata = [3, 5, 7, 9]
ydata = [2, 4, 6, 8]
plt.plot(xdata, ydata)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('Legend properties')
data1 = [1, 3, 5, 7]
data2 = [9, 7, 5, 3]
plt.plot(data1, label='Salary')
plt.plot(data2, label='Prices')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('Color properties')
xdata = [3, 5, 7, 9]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color='green')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('Axis name properties')
xdata = [3, 5, 7, 9]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('Line shape properties')
data1 = [1, 3, 5, 7]
data2 = [9, 7, 5, 3]
plt.plot(data1, color='b', label='dashed', linestyle='--')
plt.plot(data2, color='r', label='dotted', linestyle=':')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('X, Y range properties')
xdata = [15, 25, 35, 45]
ydata = [2, 4, 6, 8]
plt.plot(xdata, ydata, color='b', linestyle='--', marker='o', markersize='7')
plt.xlim(10,50)
plt.ylim(0,10)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('04_data1.xlsx', index_col='area')
data.head()
```

```python
data2020_good = data['2020_good']
data2020_good
```

```python
plt.figure(figsize=(20,4))
plt.plot(data2020_good, color='b', marker='o')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2020 Fine Dust(pm10) Good Line Graph')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,4))

for year in range(2015, 2021):
  year = str(year)+'_good'
  chartdata = data[year]
  plt.plot(chartdata, marker='o', label=year)

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015~2020 Fine Dust(pm10) Good Line Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('04_data1.xlsx', index_col='area')
data2015_verybad = data['2015_verybad']

plt.figure(figsize=(20,4))
plt.bar(data2015_verybad.index, data2015_verybad, color='g')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015 Fine Dust(pm10) Very Bad Bar Graph')
plt.grid()
plt.show()
```

```python
import numpy as np
index = np.arange(7)

plt.figure(figsize=(20,4))

data6 = data.loc['Seoul':'Ulsan', '2015_good':'2020_good']
for year in range(2015, 2021):
  year = str(year)+'_good'
  chartdata = data6[year]
  plt.bar(index, chartdata, width=0.15, label=year)
  index = index + 0.15

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.xticks(index-0.53, ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan'])
plt.title('2015~2020 Fine Dust(pm10) Good Group Bar Graph')
plt.legend()
plt.show()
```

```python
import numpy as np
index = np.arange(7)

plt.figure(figsize=(10,4))

data6 = data.loc['Seoul':'Ulsan', '2017_common':'2020_common']
for year in range(2017, 2021):
  year = str(year)+'_common'
  chartdata = data6[year]
  plt.barh(index, chartdata, label=year)

plt.xlim(120,310)
plt.ylabel('Area')
plt.xlabel('Micrometer')
plt.yticks(index, ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan'])
plt.title('2017~2020 Fine Dust(pm10) Common Group Barh Graph')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('04_data2.csv')
data
```

```python
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']

plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth)

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Iris Petal Scatter Graph')
plt.grid()
plt.show()
```

```python
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']

SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']

plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth, color='r', edgecolor='w', label='PetalLength * PetalWidth')
plt.scatter(SepalLength, SepalWidth, color='b', edgecolor='w', label='SepalLength * SepalWidth')

plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Iris Petal Group Scatter Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('04_data2.csv')
data
```

```python
setosa = data.loc[data.Species=='setosa',['Species','PetalLength']]

plt.figure(figsize=(10,6))
plt.hist(setosa['PetalLength'], bins=8, label='Petal Length')

plt.xlabel('Petal')
plt.ylabel('Length')
plt.title('Iris Setosa Petal Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
versicolor_PetalLength = data.loc[data.Species=='versicolor',['Species','PetalLength']]
virginica_PetalLength = data.loc[data.Species=='virginica',['Species','PetalLength']]

plt.figure(figsize=(10,6))
plt.hist(versicolor_PetalLength['PetalLength'], bins=5, alpha=0.5, label='versicolor')
plt.hist(virginica_PetalLength['PetalLength'], bins=5, alpha=0.5, label='virginica')

plt.xlabel('Petal')
plt.ylabel('Length')
plt.title('Iris Petal Group Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']

plt.figure(figsize=(8,5))
plt.boxplot([SepalLength, SepalWidth], labels=['Sepal Length','Sepal Width'])

plt.xlabel('Sepal')
plt.ylabel('Value')
plt.title('Iris Sepal Box Plot')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot([SepalLength, SepalWidth], labels=['Sepal Length','Sepal Width'], vert=False)

plt.ylabel('Sepal')
plt.xlabel('Value')
plt.title('Iris Sepal Horizontal Box Plot')
plt.show()
```
