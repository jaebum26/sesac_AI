## 4.3 시각화 라이브러리I 연습 문제

###문제1

#####기술적인 데이터 분석도 중요하지만 분석 결과 기반의 가치 창출이 데이터 분석의 핵심이다. 가치 창출을 위해서는 데이터 시각화를 적극적으로 활용하면 된다. 데이터 분석 역량은 데이터 수집, 가공, 분석 기법을 활용한 데이터 분석 기술이 필요하다. 반면 데이터 시각화에는 데이터 분석 결과를 시각적으로 표현해서 스토리텔링 하는 역량이다. 최근에는 데이터 시각화 활용이 쉬워지면서 고도의 전문 역량이 없더라도 데이터를 활용하여 인사이트를 발견하고, 스토리텔링을 할 수 있게 되었다. 이러한 데이터 시각화의 장점은 다음과 같다.

- 많은 양의 데이터를 시각적으로 요약
- 시각을 통한 데이터 인사이트 도출
- 3) 더 정확한 데이터 분석 결과 도출

###문제2

- 맷플롯립(Matplotlib) : 파이썬에서 가장 많이 사용하며, 데이터를 차트나 플롯으로 시각화하는 라이브러리이다. 판다스의 데이터프레임을 바로 시각화 할 때도 내부적으로 맷플롯립을 사용한다. 맷플롯립은 데이터 분석 이전의 데이터 이해를 위한 시각화 또는 데이터 분석 이후의 결과를 시각화하기 위해 사용된다.
- 시본(Seaborn) : 맷플롯립을 기반으로 색 테마, 차트 기능 등을 추가해 주는 라이브러리이다. 맷플롯립과 함께 사용하며 히트맵, 카운트플롯 등을 제공한다.
- 폴리움(Folium) : 지도 데이터를 이용하여 위치 정보를 시각화하는 라이브러리이다. 자바스크립트 기반으로 상호작용의 그래프를 그릴 수 있다.
- 파이차트(Pyecharts) : 바이두(Baidu)에서 데이터 시각화를 위해 만든 파이썬 버전의 라이브러리이다. 다양한 그래프들이 내장되어 있고 자바스크립트 기반으로 상호작용의 그래프를 그릴 수 있다.
- 플로트나인(Plotnine) : R의 ggplot2에 기반을 두어 그래프를 그려주는 라이브러리이다. R 시각화 경힘이 있다면 편리하게 사용할 수 있다.
- 플로틀리(Plotly) : 상호작용의 그래프를 그려주는 라이브러리이다. R, 스칼라, 파이썬, 자바스크립트, 매트랩 등에서 사용할 수 있다. 사용이 쉽고 세련된 도구이며 온라인과 오프라인 버전이 따로 존재한다.

###문제3

#####판다스의 시리즈와 데이터프레임에는 plot() 이라는 시각화 함수를 내장하고 있다. plot() 함수도 맷플롯립을 임포트(import)하여 사용해야 한다. 일반적인 시각화 단계를 정리하면 다음과 같다.

- 1단계: 시각화 라이브러리 불러오기
- 2단계: x축, y축에 표시할 데이터 정하기
- 3단계: plot() 함수에 데이터 입력하기
- 4단계: 그래프 보여주기

###문제4

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('04_data2.csv')
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']
SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']
```

```python
plt.figure(figsize=(15,4))
plt.plot(PetalLength, color='r', marker='o', label='Petal Length')
plt.plot(PetalWidth, color='g', marker='o', label='Petal Width')
plt.plot(SepalLength, color='b', marker='o', label='Sepal Length')
plt.plot(SepalWidth, color='y', marker='o', label='Sepal Width')
plt.xlabel('Iris')
plt.ylabel('Value')
plt.title('Iris Line Graph')
plt.legend()
plt.grid()
plt.show()
```

###문제5

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('04_data2.csv')
PetalLength = data['PetalLength'].mean()
PetalWidth = data['PetalWidth'].mean()
SepalLength = data['SepalLength'].mean()
SepalWidth = data['SepalWidth'].mean()
```

```python
plt.figure(figsize=(10,5))
plt.bar('Petal Length', PetalLength)
plt.bar('Petal Width', PetalWidth)
plt.bar('Sepal Length', SepalLength)
plt.bar('Sepal Width', SepalWidth)
plt.xlabel('Iris')
plt.ylabel('Value')
plt.title('Iris Bar Graph')
plt.grid()
plt.show()
```

###문제7

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('04_data2.csv')
```

```python
virginica = data.loc[data.Species=='virginica',['Species','SepalLength']]

plt.figure(figsize=(10,6))
plt.hist(virginica['SepalLength'], bins=7, label='Sepal Length')

plt.xlabel('Sepal')
plt.ylabel('Length')
plt.title('Iris Virginica Sepal Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('04_data2.csv')
```
