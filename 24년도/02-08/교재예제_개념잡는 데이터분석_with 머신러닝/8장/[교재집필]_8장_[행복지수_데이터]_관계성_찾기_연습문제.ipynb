## 연관분석

```python
# 필요한 pandas, mlxtend 패키지를 로드한다
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
# 데이터셋 생성하기
dataset=[['치즈','우유','생수'],
['생수','수박','우유','맥주'],
['수박','생수','맥주'],
['생수','우유','맥주'],
['치즈','우유','맥주']
]
```

```python
# 문제7-1)  다음은 생성된 데이터셋을 통해 학습을 시작하는 코드이다. 실행결과를 보고 빈 칸에 가장 알맞은 코드를 완성하시오. 
te = TransactionEncoder()
te_arr = te.fit(dataset).transform(dataset)
te_arr
```

```python
# 문제7-2)  다음은 7-1에서 생성한 배열을 데이터프레임으로 변경하는 코드이다. 실행결과를 보고 빈 칸에 가장 알맞은 코드를 완성하시오.
df=pd.DataFrame(te_arr,columns=te.columns_)
df
```

```python
# # 문제7-3)  다음은 7-2에서 생성한 데이터프레임으로 apriori 알고리즘 사용하여 지지도(support)를 알아보는 코드이다. 실행 결과를 보고 빈 칸에 가장 알맞은 코드를 완성하시오. 최소지지도는 0.5로 설정한다.
itemset = apriori(df, min_support=0.5, use_colnames=True)
itemset
itemset.sort_values('support',ascending=False)
# 특별히 옵션을 주지 않으면 기본 지지도(support)는 0.5로 설정된다. 
```

```python

itemset = apriori(df, min_support=0.4, use_colnames=True)
itemset.sort_values('support',ascending=False).head()
# 생수를 살 확률은 100%, 우유와 생수를 살 확률은 75%
```

```python
# 신뢰도 확인
# min_threshold 의 기본값은 0.8 이다.
```

```python
#문제7-4)  다음은 7-3 결과를 이용하여 연관규칙을 알아보는 코드이다. 실행 결과를 보고 빈 칸에 가장 알맞은 코드를 완성하시오. 최소신뢰도는 0.8로 설정(기본값)한다.
from mlxtend.frequent_patterns import association_rules
association_rules(itemset, metric='confidence',min_threshold=0.8) 
# 5번 인덱스 데이터를 보면 (치즈,생수)를 구매한 고객이 (우유)를 구매할  확률이 높다고 판단 할 수 있다.
```
