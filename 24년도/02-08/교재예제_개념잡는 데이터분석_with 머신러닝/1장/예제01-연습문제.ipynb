#####문제 해결 또는 인사이트 도출을 위한 목적으로 데이터 수집, 전처리 및 정형화, 분석과 표현하는 과정이다. 빅데이터 분석은 여러 가지 접근 방식이 있으며 다양한 기술을 조합하여 경제, 과학, 사회과학 등 다양한 분야에서 활용되고 있다. 오늘날 비즈니스 분야에서 빅데이터 분석은 의사 결정을 더욱더 과학적이고 효율적으로 운영할 수 있도록 도와주는 중요한 도구로 활용되고 있다.

###문제2 - 빅데이터 분석을 위한 데이터 특징을 작성하시오.

- 문제 해결 또는 인사이트 도출을 위한 데이터
#####빅데이터 분석은 문제 해결이나 개선 방안을 찾기 위함이다. 목적에 맞지 않은 데이터는 오히려 빅데이터 분석에 방해가 된다. 예를 들어 근육질의 건강한 몸매를 만들기 위해 학력이나 정치 성향 정보는 필요하지 않다. 상관없는 데이터는 빅데이터 분석을 복잡하게 만들기 때문에 반드시 필요한 데이터인지를 잘 판단해야 한다.

- 정확한 수치 데이터
#####빅데이터 분석을 위해서는 수집한 데이터가 정확한 수치 데이터여야 하고 그 수치 표현 형식도 일관되어야 한다. 예를 들어 10년 전에 측정한 체중 또는 킬로그램(kg)이나 파운드(lb) 단위를 혼용해서 작성한다면 빅데이터 분석 과정에서 오류가 발생하거나 복잡한 상황이 발생할 것이다. 빅데이터 분석을 위한 데이터는 정확하고 일관되어야 결과도 정확하게 된다.

- 적당한 양의 데이터
#####빅데이터 분석을 할 때 데이터가 너무 적으면 한쪽으로 편향된 결과가 나타나 문제 해결 및 인사이트 도출에 적절하지 않은 정보를 제시할 수도 있다. 반면 데이터가 너무 많으면 복잡하고 데이터 자체의 가치와 일관성이 떨어질 수도 있다. 빅데이터 분석을 위해서는 적당한 양의 데이터가 필요한데, 적당한 양의 기준은 빅데이터 분석 목적에 따라 결정된다.
