```python
import requests
import pandas as pd
import re

def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

def get_naver_shopping_data(client_id, client_secret, query, display_count=5, max_price=1000000):
    # 네이버 쇼핑 API 엔드포인트
    api_url = "https://openapi.naver.com/v1/search/shop.json"
    # API 요청 헤더 설정
    headers = {
        "X-Naver-Client-Id": client_id,
        "X-Naver-Client-Secret": client_secret
    }
    # API 요청 파라미터 설정
    params = {
        "query": query,
        "display": display_count,
        "maxPrice": max_price,
        "start": 1  # 검색 시작 위치 (1부터 시작)
    }
    # API 요청 보내기
    response = requests.get(api_url, headers=headers, params=params)
    # 응답 확인
    if response.status_code == 200:
        # JSON 형식의 응답을 데이터프레임으로 변환하여 반환
        result = response.json()
        df = pd.DataFrame(result['items'])
        # HTML 태그 제거
        df['title'] = df['title'].apply(remove_html_tags)
        # 가격이 0인 제품 및 가격이 100만원을 초과하는 제품 제거
        df = df[(df['lprice'] != '0') & (df['lprice'].astype(int) <= max_price)]
        return df
    else:
        # 에러 발생 시 에러 코드 출력
        print(f"Error: {response.status_code}, {response.text}")
        return None

# 네이버 API 클라이언트 ID와 클라이언트 시크릿 입력
client_id = "kdDlQECbpJx57H36m3qF"
client_secret = "9gPfy4kdmX"
# 검색할 키워드 입력
search_query = "삼성전자 노트북"
display_count = 50
max_price = 1000000  # 100만원 이하로 설정
# 제품 데이터를 데이터프레임으로 가져오기
result_df = get_naver_shopping_data(client_id, client_secret, search_query, display_count, max_price)
# 결과 출력
if result_df is not None:
    print(result_df)

result_df.head()
```

```python
import requests

def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

def get_naver_shopping_data(client_id, client_secret, query, display_count=5):
    api_url = "https://openapi.naver.com/v1/search/shop.json"
    headers = {
        "X-Naver-Client-Id": client_id,
        "X-Naver-Client-Secret": client_secret
    }
    params = {
        "query": query,
        "display": display_count,
        "start": 1
    }

    response = requests.get(api_url, headers=headers, params=params)

    if response.status_code == 200:
        result = response.json()
        items = result.get('items', [])

        # HTML 태그 제거
        for item in items:
            item['title'] = remove_html_tags(item.get('title', ''))

        return items
    else:
        print(f"Error: {response.status_code}, {response.text}")
        return None

def find_lowest_price(items):
    lowest_price_info = None

    for item in items:
        title = item.get('title', '')
        link = item.get('link', '')
        price = int(item.get('lprice', 0))

        if lowest_price_info is None or price < lowest_price_info['price']:
            lowest_price_info = {
                'title': title,
                'price': price,
                'link': link
            }

    return lowest_price_info

# 검색할 키워드 입력
search_query = "LG스마트TV"
display_count = 100

# 네이버 쇼핑 API를 통해 데이터 가져오기
lg_tv_items = get_naver_shopping_data(client_id, client_secret, search_query, display_count)

if lg_tv_items:
    # 최저가 정보 찾기
    lowest_price_info = find_lowest_price(lg_tv_items)

    if lowest_price_info:
        print("최저가 상품 정보:")
        print(f"상품명: {lowest_price_info['title']}")
        print(f"가격: {lowest_price_info['price']}원")
        print(f"쇼핑몰 링크: {lowest_price_info['link']}")
    else:
        print("가격 정보를 찾을 수 없습니다.")
else:
    print("데이터를 가져오는데 실패했습니다.")
```

```python
import requests
import os
import random

def download_images(query, api_key, output_dir, num_images=1):
    endpoint = "https://www.googleapis.com/customsearch/v1"
    cx = "5326759e7cb864e89"  # Google Custom Search Engine ID
    
    # 페이지당 결과 수를 1로 설정
    start = 1
    num = 1

    params = {
        "q": query,
        "cx": cx,
        "key": api_key,
        "searchType": "image",
        "start": start,
        "num": num
    }

    response = requests.get(endpoint, params=params)
    
    if response.status_code == 200:
        data = response.json()
        if 'items' in data:
            # 이미지를 저장할 디렉토리 생성
            if not os.path.exists(output_dir):
                os.makedirs(output_dir)

            for i, item in enumerate(data['items']):
                image_url = item['link']
                image_extension = os.path.splitext(image_url)[-1]
                image_path = os.path.join(output_dir, f"{query}_{i}{image_extension}")

                # 이미지 다운로드
                image_response = requests.get(image_url)
                with open(image_path, 'wb') as f:
                    f.write(image_response.content)
                    print(f"다운로드 완료: {image_path}")
        else:
            print("이미지를 찾을 수 없습니다.")
    else:
        print(f"API 요청이 실패했습니다. 에러 코드: {response.status_code}, 메시지: {response.text}")

# 사용 예시
query = "고양이"  # 원하는 이미지를 검색할 쿼리
api_key = "AIzaSyBjFBGivCJ3bVZGSEqEq9RaAFOW6xqLz3E"  # 발급받은 Google Custom Search API 키
output_directory = "downloaded_images"  # 이미지를 저장할 디렉토리

# 이미지 다운로드
download_images(query, api_key, output_directory)
```

```python
import requests

def geocode_address(api_key, address):
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    # API 요청 파라미터 설정
    params = {
        'address': address,
        'key': api_key,
    }

    # API 요청 보내기
    response = requests.get(base_url, params=params)
    result = response.json()

    # 응답 결과 확인
    if response.status_code == 200:
        if result['status'] == 'OK' and len(result['results']) > 0:
            location = result['results'][0]['geometry']['location']
            latitude = location['lat']
            longitude = location['lng']
            return latitude, longitude
        else:
            print("주소를 찾을 수 없습니다.")
            return None
    else:
        print(f"API 요청이 실패했습니다. 에러 코드: {response.status_code}, 메시지: {result.get('error_message')}")
        return None

# 사용 예시
api_key = "AIzaSyAnTV4bT4oGURXZXASCEyqvCCGk50Qa_Qo"  # 본인의 Google Cloud Console에서 발급받은 API 키로 대체
address = "서울시 강남구"

result = geocode_address(api_key, address)

if result:
    print(f"{address}의 위경도: {result}")
```
