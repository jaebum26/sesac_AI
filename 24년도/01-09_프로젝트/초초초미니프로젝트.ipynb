```python
import cv2
import mediapipe as mp
import numpy as np
import time
import subprocess
import pyautogui

# 가상 키보드 실행
subprocess.Popen("osk.exe", shell=True, stdin=subprocess.PIPE, stdout=subprocess.PIPE)

# MediaPipe의 키 관절 모델 로드
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=False, max_num_hands=2, min_detection_confidence=0.5)

# OpenCV 비디오 캡처 초기화 및 해상도 설정
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

# 키보드 버튼 중앙 좌표 및 크기 설정 (영어 키보드 기준)
keyboard_buttons = {
    (50, 150, 100, 200): '1', (150, 150, 200, 200): '2', (250, 150, 300, 200): '3', (350, 150, 400, 200): '4', (450, 150, 500, 200): '5', (550, 150, 600, 200): '6', (650, 150, 700, 200): '7', (750, 150, 800, 200): '8', (850, 150, 900, 200): '9', (950, 150, 1000, 200): '0',
    (50, 250, 100, 300): 'q', (150, 250, 200, 300): 'w', (250, 250, 300, 300): 'e', (350, 250, 400, 300): 'r', (450, 250, 500, 300): 't', (550, 250, 600, 300): 'y', (650, 250, 700, 300): 'u', (750, 250, 800, 300): 'i', (850, 250, 900, 300): 'o', (950, 250, 1000, 300): 'p',
    (50, 350, 100, 400): 'a', (150, 350, 200, 400): 's', (250, 350, 300, 400): 'd', (350, 350, 400, 400): 'f', (450, 350, 500, 400): 'g', (550, 350, 600, 400): 'h', (650, 350, 700, 400): 'j', (750, 350, 800, 400): 'k', (850, 350, 900, 400): 'l',
    (100, 450, 150, 500): 'z', (200, 450, 250, 500): 'x', (300, 450, 350, 500): 'c', (400, 450, 450, 500): 'v', (500, 450, 550, 500): 'b', (600, 450, 650, 500): 'n', (700, 450, 750, 500): 'm',
    (950, 350, 1080, 400): 'Shift', (800, 450, 950, 500): 'Enter',
    (1050, 150, 1230, 200): 'Backspace',  # 백스페이스 버튼
    (300, 550, 700, 600): 'Space'  # 스페이스 버튼 (넓게 구현)
    # ... 필요한 만큼 더 많은 키를 추가하십시오
}

typed_text = ""  # 입력된 텍스트를 저장할 변수
mark = {
    'start': 8,
    'end': 6
}
last_key_press_time = 0  # 마지막으로 키를 누른 시간
shift_pressed = False  # 쉬프트 키 상태를 추적

while cap.isOpened():
    ret, frame = cap.read()
    # 좌우 반전
    frame = cv2.flip(frame, 1)
    if not ret:
        break

    # 배경 이미지 생성
    background = np.zeros_like(frame)

    for button_coords, button_char in keyboard_buttons.items():
        x1, y1, x2, y2 = button_coords
        cv2.rectangle(background, (x1, y1), (x2, y2), (0, 255, 255), -1)
        # 수정: 쉬프트 눌렸을 때 소문자와 대문자를 서로 바꿔서 표시
        if shift_pressed:
            button_char_display = button_char.upper()
        else:
            button_char_display = button_char.lower()
        cv2.putText(background, button_char_display, (x1 + 10, y1 + 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

    # 키보드 이미지와 현재 프레임을 섞어줌
    alpha = 0.6  # 투명도 조절
    frame = cv2.addWeighted(frame, alpha, background, 1 - alpha, 0)

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(rgb_frame)
    current_time = time.time()

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            for idx, landmark in enumerate(hand_landmarks.landmark):
                if idx == mark['start'] or idx == mark['end']:
                    landmark_x = int(landmark.x * 1280)  # 새로운 해상도로 조정
                    landmark_y = int(landmark.y * 720)   # 새로운 해상도로 조정
                    cv2.circle(frame, (landmark_x, landmark_y), 5, (255, 0, 0), -1)

            landmark5_x = int(hand_landmarks.landmark[mark['start']].x * 1280)  # 새로운 해상도로 조정
            landmark5_y = int(hand_landmarks.landmark[mark['start']].y * 720)   # 새로운 해상도로 조정
            landmark7_x = int(hand_landmarks.landmark[mark['end']].x * 1280)    # 새로운 해상도로 조정
            landmark7_y = int(hand_landmarks.landmark[mark['end']].y * 720)     # 새로운 해상도로 조정

            if abs(landmark5_x - landmark7_x) < 20 and abs(landmark5_y - landmark7_y) < 20 and (
                    current_time - last_key_press_time) > 1.0:
                for button_coords, button_char in keyboard_buttons.items():
                    x1, y1, x2, y2 = button_coords
                    if x1 < landmark5_x < x2 and y1 < landmark5_y < y2:
                        if button_char == 'Shift':
                            shift_pressed = not shift_pressed
                        elif button_char == 'Enter':
                            pyautogui.press('enter')
                        elif button_char == 'Backspace':
                            # 백스페이스 동작
                            pyautogui.press('backspace')
                            typed_text = typed_text[:-1]
                        elif button_char == 'Space':
                            # 스페이스 동작
                            typed_text += ' '
                            pyautogui.press('space')
                        else:
                            # 쉬프트 키가 눌려 있으면 대문자로 변환
                            if shift_pressed:
                                button_char = button_char.upper()
                            typed_text += button_char
                            # 해당 키에 대응하는 동작 실행
                            pyautogui.press(button_char)
                            last_key_press_time = current_time

    cv2.putText(frame, "Typed Text: " + typed_text, (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Virtual Keyboard', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
 
cap.release()
cv2.destroyAllWindows()
```
