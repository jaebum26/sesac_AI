```python
from konlpy.tag import Okt
import pandas as pd

# 한글 형태소 분석기 초기화
okt = Okt()

# 빈 데이터프레임 생성
combined_df = pd.DataFrame(columns=['이유', 'str'])

# 주어진 데이터프레임의 특정 범위에 대해 작업 수행
for i in range(2000):  # 0부터 1000까지의 인덱스를 반복
    text = df['판례내용'][i]  # 해당 인덱스의 '판례내용' 텍스트를 가져옴
    text_f = df['사건종류명'][i]
    text_n = df['법원명'][i]
    
    # '이유'라는 단어의 인덱스 찾기
    reason_index = text.find('【이    유】')
    
    # '이유' 이후의 내용 가져오기
    reason_text = text[reason_index + len('【이    유】'):]
    
    # 명사만 추출하여 토큰화
    nouns = okt.nouns(reason_text)
    nouns_text = ' '.join(nouns)
    
    # 추출된 내용을 데이터프레임에 추가
    combined_df = combined_df.append({'법원명' : text_n, '사건종류명' : text_f, '판례내용': text, '이유': reason_text, 'str': nouns_text}, ignore_index=True)

# CSV 파일로 저장
combined_df.to_csv('reason_text.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df_r = pd.read_csv('reason_text.csv')
df_r.head()
```

```python
# 판례내용이 저장된 CSV 파일 읽기
df = pd.read_csv('reason_text.csv')

# TF-IDF 매트릭스 생성
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['str'])

def recommend_cases(question, top_n=1):
    # 입력 질문에 대한 TF-IDF 벡터 생성
    question_tfidf = tfidf_vectorizer.transform([question])
    
    # 입력 질문과 판례내용 간의 코사인 유사도 계산
    similarities = cosine_similarity(question_tfidf, tfidf_matrix)
    
    # 코사인 유사도를 기준으로 가장 유사한 판례 인덱스 추출
    top_similar_cases_indices = similarities.argsort(axis=1)[:, -top_n:][0][::-1]
    
    # 추천된 판례의 판례내용 출력
    recommended_cases = df.iloc[top_similar_cases_indices]['판례내용']
    return recommended_cases

# 예시: 질문에 대한 판례 추천
question = input('질문에 대한 판례를 추천해드립니다 : ')
recommended_cases = recommend_cases(question)
for case in recommended_cases:
    print(case)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['font.family'] = 'NanumGothic'  # 한글 폰트 지정

# 네트워크 그래프 생성
G = nx.Graph()

# '사건종류명'이 '형사'인 경우에 대해서만 노드 추가
nodes = []
for idx, row in df.iterrows():
    if '형사' in row['사건종류명']:
        G.add_node(row['법원명'])
        nodes.append(row['법원명'])

# 각 노드의 빈도 계산
node_counts = {node: nodes.count(node) for node in G.nodes()}

# 노드 크기 설정
node_sizes = [node_counts[node] * 100 for node in G.nodes()]

# 네트워크 그래프 시각화
plt.figure(figsize=(10, 10))
nx.draw(G, with_labels=True, node_size=node_sizes, font_family='Malgun Gothic', font_size=12)
plt.show()
```

```python
import pandas as pd
import networkx as nx
from itertools import combinations
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('reason_text.csv')

# 네트워크 그래프 생성
G = nx.Graph()

# 각 텍스트에서 등장하는 단어들을 노드로 추가
for idx, row in df.iterrows():
    words = row['str'].split()  # 공백을 기준으로 단어들을 분리
    G.add_nodes_from(words)

# 각 텍스트에서 등장하는 단어들의 조합을 엣지로 추가
for idx, row in df.iterrows():
    words = row['str'].split()  # 공백을 기준으로 단어들을 분리
    for word1, word2 in combinations(words, 2):  # 단어들의 모든 조합에 대해
        if G.has_edge(word1, word2):  # 이미 엣지가 있는 경우 가중치를 증가
            G[word1][word2]['weight'] += 1
        else:  # 엣지가 없는 경우 새로운 엣지 추가
            G.add_edge(word1, word2, weight=1)

# 그래프 시각화
plt.figure(figsize=(15, 10))
pos = nx.spring_layout(G, k=0.1)  # 그래프 레이아웃 정의
nx.draw(G, pos, node_size=100, with_labels=False, font_size=8)  # 노드 및 엣지 그리기
plt.show()
```
